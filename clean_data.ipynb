```python
import pandas as pd




# Get CSV files and fill na values with -1
df = pd.read_csv("all_data.csv", sep="~")
df = df.fillna("-1")
new_data = {}
print(df)
```

```python
#extract features from dataframe
values = df.loc[:, [str(i) for i in range(3, 58)]]
new_df = []
count = 0
for i in range(len(values)):
    new_row = []
    for j in range(len(values.columns)):
        d = values.iloc[i, j]
        if d == "-1":
            continue
        d = d.split("||")
        new_row += d
    new_df.append(new_row)
```

```python
#function to refactor data into specific categories
# categories: what to look for (["strings", "strings", ...])
# iterations: how many rows to refactor (int)
def extract_data(categories=["architectural style", "bathrooms", "bedrooms", "date on market", "size", "home type", "year built", "sewer", "water"], iterations=0):
    for cat in categories:
        col = []
        for i in range(iterations):
            data = new_df[i]
            has_feature = False
            for feature in data:
                if feature.startswith(cat + ":"):
                    f = feature.split(": ")
                    col.append(f[1])
                    has_feature = True
                    break #for now stop after finding it once
            if not(has_feature):
                col.append("Nan")
        new_data[cat] = col
```

```python
# run and create a new dataframe
# CREATE YOUR OWN CATEGORIES HERE
categories = []
extract_data(categories, len(values))

df_features = pd.DataFrame(new_data)
df = df.drop(columns = [str(i) for i in range(3, 58)])

df = pd.concat([df_features, df], axis=1, ignore_index=True)
```

```python
#rename columns, drop unnecessary columns (optional)
df = df.drop(columns = [10, 11])
df.columns = ["architectural style", "bathrooms", "bedrooms", "date on market", "size", "home type", "year built", "sewer", "water", "imgs", "price", "address", "link1", "link2"]
print(df)

df.to_csv("final_data.csv", sep = "~")
```
